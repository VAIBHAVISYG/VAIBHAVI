```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
```

```python
stroke = pd.read_csv('healthcare-dataset-stroke-data_vaibhavi1.csv')
```

```python
stroke.info()
```

```python
stroke.isnull().sum()
```

```python
stroke["bmi"].fillna( method ='ffill', inplace = True)
```

```python
stroke.duplicated()
```

```python
sns.boxplot(data = stroke['bmi'], orient = 'v', color = 'b')
```

```python
q1, q2 =np.percentile(stroke['bmi'],[25,75])
```

```python
print(q1,q2)
```

```python
iqr = q2-q1
iqr
```

```python
lower_fence = q1-(1.5*iqr)
upper_fence = q2+(1.5*iqr)
print(lower_fence)
print(upper_fence)
```
